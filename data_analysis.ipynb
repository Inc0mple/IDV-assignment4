```python
import pandas as pd

# Load the data from CSV
# In a real scenario, you would replace this with the actual file path
df = pd.read_csv('data/df3_sg_residents_by_age_and_sex.csv')

# Filter for rows where both Age and Sex are "Total"
# This gives us the total population for each planning area and subzone
total_population = df[(df['Age'] == 'Total') & (df['Sex'] == 'Total')]

# Group by Planning Area and Subzone, and sum the 2024 population
subzone_totals = total_population.groupby(['Planning Area', 'Subzone'])['2024'].sum().reset_index()

# Sort by Planning Area and then by Subzone
subzone_totals = subzone_totals.sort_values(['Planning Area', 'Subzone'])

# Rename the column for clarity
subzone_totals = subzone_totals.rename(columns={'2024': 'Total Population'})

# Display the results
print(subzone_totals)

# Optionally, save the results to a new CSV file
subzone_totals.to_csv('population_by_subzone.csv', index=False)
```
